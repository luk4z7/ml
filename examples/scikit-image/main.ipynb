```python
# Carregando imagens default da library skimage
from skimage import data
astronaut = data.astronaut()
astronautOrigin = data.astronaut()
imshow(astronaut)
```

```python
astronaut.shape
```

```python
# Terceira dimensão representa as cores
astronaut[0, 0]
```

```python
imshow(astronaut[400:, 300:, :])
```

```python
# Obtendo somente o primeiro canal 
astronaut[:, :, 0] = 0 # zerando os valores do vermelho
astronaut[:, :, 1] = 0 # zerando os valores do verde
# Imagem somente com os valores dos azuis
imshow(astronaut)
```

```python
from skimage.filters import gaussian
```

```python
# Efetuando uma cópia sem usar a referência de atribuição de váriaveis
astronaut_copy = data.astronaut().copy()
imshow(gaussian(astronaut_copy, sigma=16))
```

```python
from skimage.io import imsave
# Salvando uma imagenm
imsave('astronaut.png', astronaut_copy)
```

```python
# Aplicando um noise na imagem
from skimage.util import random_noise
astronaut_noise = random_noise(astronaut_copy)
imshow(astronaut_noise)
```

```python
# Removendo o noise com gaussian
imshow(gaussian(astronaut_noise, sigma=1))
```

```python
from skimage.restoration import denoise_bilateral
# Removendo o noise com denoise_bilateral
imshow(denoise_bilateral(astronaut_noise, sigma_spatial=1))
```
